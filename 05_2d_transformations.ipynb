# Importación de Librerías

```python
import requests
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Lectura y Despliegue de Imagen


```python
image_url = 'https://raw.githubusercontent.com/bartolomeoadrian/computational-vision/refs/heads/main/assets/images/iguazu.jpg'

# Download the image
response = requests.get(image_url)
with open('iguazu.jpg', 'wb') as f:
    f.write(response.content)

# Read the image using cv2.imread
img = cv2.imread('iguazu.jpg')

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols = img_rgb.shape[:2]

plt.imshow(img_rgb)
plt.show()
```

# Traslación

```python
dx, dy = 200, 200
M_trans = np.float32([[1, 0, dx],
                      [0, 1, dy]])

translated = cv2.warpAffine(img_rgb, M_trans, (cols, rows))

plt.imshow(translated)
plt.show()
```

# Rotación

```python
angle = 45
center = (cols // 2, rows // 2)
M_rot = cv2.getRotationMatrix2D(center, angle, 1.0)

rotated = cv2.warpAffine(img_rgb, M_rot, (cols, rows))

plt.imshow(rotated)
plt.show()
```

# Escalamiento

```python
scale_x, scale_y = 1, 0.5
M_scale = np.float32([[scale_x, 0, 0],
                      [0, scale_y, 0]])

scaled = cv2.warpAffine(img_rgb, M_scale, (int(cols * scale_x), int(rows * scale_y)))

plt.imshow(scaled)
plt.show()
```
