# Importación de Librerías

```python
import requests
import numpy as np
from cv2 import imread
import matplotlib.pyplot as plt
```

# Lectura y Despliegue de Imagen

```python
image_url = 'https://raw.githubusercontent.com/bartolomeoadrian/computational-vision/5adc2f404c6bcbb468a57babec3a99a242bd10c3/assets/images/moon.jpg'

# Download the image
response = requests.get(image_url)
with open('moon.jpg', 'wb') as f:
    f.write(response.content)

# Read the image using cv2.imread
img = imread('moon.jpg')
plt.imshow(img)
plt.show()
```

# Selección del primer canal de la imagen

```python
one_channel_image = img[:,:,0]
plt.imshow(one_channel_image)
plt.show()
```

# Elección del umbral de segmentación


```python
plt.hist(one_channel_image.ravel(), bins=256, range=[0,256])
plt.title('Histogram of Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

# Segmentación por Umbral

```python
def segmenta(image,t):
  (N,M) = image.shape
  Y = np.zeros((N,M))
  for i in range(N):
    for j in range(M):
      if image[i,j] > t:
        Y[i,j] = 255
  return Y

Y = segmenta(one_channel_image,35)
plt.imshow(Y)
plt.show()
```
